```python
import yasa
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
openResting = pd.read_csv('../data/openResting.csv')
openPedalling = pd.read_csv('../data/openPedalling.csv')
closedResting = pd.read_csv('../data/closedResting.csv')
closedPedalling = pd.read_csv('../data/closedPedalling.csv')
```

```python
data = pd.Series(openResting['amplitude'].to_numpy(),
                 index = openResting['electrode'],
                 name = 'Entropy')
fig_openResting = yasa.topoplot(data, vmin=4.5, vmax=5.5, cmap = 'flare', title='Resting Open', figsize = (5, 5), dpi = 600, fontsize = 18)
plt.savefig('openResting_head.pdf', dpi=600)
```

```python
data = pd.Series(openPedalling['amplitude'].to_numpy(),
                 index = openResting['electrode'],
                 name = 'Entropy')
fig_openPedalling = yasa.topoplot(data, vmin=4.5, vmax=5.5, cmap = 'flare', title='Pedalling Open', figsize = (5, 5), dpi = 600, fontsize = 18)
plt.savefig('openPedalling_head.pdf', dpi=600)
```

```python
data = pd.Series(closedResting['amplitude'].to_numpy(),
                 index = openResting['electrode'],
                 name = 'Entropy')
fig_closedResting = yasa.topoplot(data, vmin=4.5, vmax=5.5, cmap = 'flare', title='Resting Closed', figsize = (5, 5), dpi = 600, fontsize = 18)
plt.savefig('closedResting_head.pdf', dpi=600)
```

```python
data = pd.Series(closedPedalling['amplitude'].to_numpy(),
                 index = openResting['electrode'],
                 name = 'Entropy')
fig_closedPedalling = yasa.topoplot(data, vmin=4.5, vmax=5.5, cmap = 'flare', title='Pedalling Closed', figsize = (5, 5), dpi = 600, fontsize = 18)
plt.savefig('closedPedalling.pdf', dpi=600)
```
